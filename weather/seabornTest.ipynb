```python
import matplotlib.pyplot as plt
summary=df.groupby("HODESH_TEUNA").agg(num_of_accidents=("HODESH_TEUNA","size"),severirty=("SUG_TEUNA","size"))
         #size().plot(kind="bar",color='teal'))
fig, ax1=plt.subplots(figsize=(8,5))
summary['num_of_accidents'].plot(kind="line",color='skyblue', ax=ax1)
#, position=1 , width=0.3
ax1.set_ylabel('Number of accidents over time',color="blue")
ax1.tick_params(axis='y', colors='blue')
ax1=plt.subplots(figsize=(8,5))
ax2 = ax1.twinx()
summary['severity'].plot(kind="line",color='darkorange', ax=ax2, marker='o')
ax2.set_ylabel('severity',color="darkorange")
ax2.tick_params(axis='y', colors='darkorange')
#plt.title("number of accidents per month")
#plt.ylabel("Count")
ax1.set_xticklabels(summary.index, rotation=0)
plt.show()
```
